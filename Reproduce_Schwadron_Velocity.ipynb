```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
table = np.genfromtxt('PSP_SWP_SPC_L3I_209776.txt', skip_header=76, skip_footer=3)
```

```python
main = np.array(table)
```

```python
main_2 = main[:,1:]
```

```python
time_int1 = (1 / 24) * (1 / 60) * (1 / 60) * main_2[:,0]
```

```python
time_int2 = time_int1 - 323
```

```python
time_h = time_int2 * 24
```

```python
Main_fix = np.column_stack([time_h, main_2[:,1:]])
```

```python
x = Main_fix[:,0]
y1 = Main_fix[:,1]
y2 = Main_fix[:,2]
```

```python
plt.figure(figsize=(16,1))
plt.xlim(1, 11)
plt.ylim(400, 600)
plt.ylabel("$V_{R}(kms^{-1})$")

plt.scatter(x, y1, s=15)
plt.show()
```

```python
plt.figure(figsize=(16,1))
plt.xlim(1, 11)
plt.ylim(-70, 70)
plt.ylabel("$V_{T}(kms^{-1})$")

plt.scatter(x, y2, s=15)
plt.show()
```
