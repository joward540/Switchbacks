```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
table = np.genfromtxt('PSP_FLD_L2_MAG_RTN_1MIN_145910.txt', skip_header=59, skip_footer=3)
```

```python
corr_table = table.T
```

```python
A1 = corr_table
```

```python
A1_d = (1 / 24) * (1 / 60) * (1 / 60) * A1[1]
```

```python
A1_t = A1_d - 323
```

```python
A1_h = A1_t * 24
A1h = np.array(A1_h)
```

```python
A1_fix = np.column_stack((A1h,A1[2],A1[3],A1[4]))
```

```python
AA = A1_fix
n = table
x = AA[:,0]
B_R = AA[:,1]
B_T = AA[:,2]
B_N = AA[:,3]
```

```python
y1 = B_R
y2 = B_T
y3 = B_N
```

```python
plt.figure(figsize=(17,1))
plt.xlim(min(AA[:,0]), max(AA[:,0]))
plt.ylim(min(AA[:,2])-5,max(AA[:,1])+5)
plt.ylabel('nT')

plt.plot(x, y1, color='g')
plt.plot(x, y2, color='r')
plt.plot(x, y3, color='b')

plt.savefig('Panel 1 Reproduction')
plt.show()
```

```python
plt.figure(figsize=(17,1))
plt.xlim(min(AA[:,0]), max(AA[:,0]))
plt.ylim(min(AA[:,2]),20)
plt.ylabel("$B_{R}$(nT)")

plt.plot(x, y1, color='g')
plt.show()
```

```python
plt.figure(figsize=(17,1))
plt.xlim(min(AA[:,0]), max(AA[:,0]))
plt.ylim(min(AA[:,2]),20)
plt.ylabel("$|B_{R}|$(nT)")

plt.plot(x, abs(y1), color='g')
plt.show()
```
